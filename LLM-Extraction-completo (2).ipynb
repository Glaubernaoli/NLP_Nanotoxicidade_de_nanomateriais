```python
import pandas as pd

def parse_resposta(texto):
    dados = {"Composition": None, "Nanoparticle Size": None, "Toxicity": None, "Type of Cells": None}
    
    # separar em linhas
    linhas = texto.split("\n")
    
    for linha in linhas:
        linha = linha.strip()
        if not linha:
            continue
        
        # normalizar: remover ** e variantes de :
        linha = linha.replace("**", "").replace("：", ":")
        
        if ":" not in linha:
            continue
        
        chave, valor = linha.split(":", 1)
        chave = chave.strip()
        valor = valor.strip().strip("[]")
        
        # detectar a qual campo corresponde
        if "composition" in chave.lower():
            dados["Composition"] = valor
        elif "nanoparticle size" in chave.lower():
            dados["Nanoparticle Size"] = valor
        elif "toxicity" in chave.lower():
            dados["Toxicity"] = valor
        elif "type of cells" in chave.lower():
            dados["Type of Cells"] = valor
    
    return dados

```

```python
# aplicar para cada resposta
linhas_df = [parse_resposta(r) for r in outputs]

# montar dataframe
df = pd.DataFrame(linhas_df)
print(df)
df.to_excel("nanotoxicology_df.xlsx", index=False)
```
